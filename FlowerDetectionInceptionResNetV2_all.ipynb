```
import os
import json

all_flower = {"AngelaFlower", "AngelaFlowerLeaves", "SandJoy", "Overwhelm", "WhiteTanic", "PinkGragon", "PinkGragonLeaves", "PurpleBelt", "PurpleBeltLeaves", "RedDiamond", "FreedomLeaves", "Freedom", "Flavor", "GrapeIcebergLeaves", "GrapeIceberg", "ShiningDiamond", "BeeCaramelLeaves", "GoldenLeaves", "BeeCaramel", "Golden"}

sample_json = []

def write_json(flowertype):
  dst = os.path.join("/content/images/", str(flowertype))
  filenames = os.listdir(dst)
  count1 = 0
  for i in filenames:
    count1 = count1 + 1
    filenames_dst = os.path.join(dst, str(i))
    filenames_new = dst + str(count1) + ".JPG"
    os.rename(filenames_dst, filenames_new)
    print(filenames_new)
  
  os.rmdir(dst)
  dst = "/content/images/"
  filenames = os.listdir(dst)
  for i in filenames:
    sample_json.append(
        {
        'image_id': i,
        'label': flowertype
        }
    )

for i in all_flower:
  print(i)
  write_json(i)


#write_json("AngelaFlower")
#write_json("AngelaFlowerLeaves")
#write_json("SandJoy")
#write_json("Overwhelm")
#write_json("WhiteTanic")
#write_json("PinkGragon")
#write_json("PinkGragonLeaves")
#write_json("PurpleBelt")
#write_json("PurpleBeltLeaves")
#write_json("RedDiamond")
#write_json("FreedomLeaves")
#write_json("Freedom")
#write_json("Flavor")
#write_json("GrapeIcebergLeaves")
#write_json("GrapeIceberg")
#write_json("ShiningDiamond")
#write_json("BeeCaramel")
#write_json("Golden")
#write_json("BeeCaramelLeaves")
#write_json("GoldenLeaves")



with open('data.json', 'w') as outfile:
    json.dump(sample_json, outfile, indent=4, sort_keys=True)
```

```
IMAGE_DIR ='images'

from imageatm.components import DataPrep

dp = DataPrep(
    image_dir = 'images/',
    samples_file = 'data.json',
    job_dir = 'imagenette'
)

dp.run()
```

```
from imageatm.components import Training

trainer = Training(
     dp.image_dir, dp.job_dir, epochs_train_dense=3, epochs_train_all=1, batch_size=64, base_model_name="InceptionResNetV2"
)

trainer.run()
```

```
 from imageatm.components import Evaluation

e = Evaluation(image_dir=dp.image_dir, job_dir=dp.job_dir) 

e.run()
```
